```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({
    'color':['red', 'blue', 'green','green','red','blue']
})
```

```python
df
```

```python
encoder = OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df[['color']])
```

```python
encodeed_df=pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out())
```

```python
pd.concat([df, encodeed_df], axis=1)
```

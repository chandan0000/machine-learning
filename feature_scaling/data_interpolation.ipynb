1. Linear interpolation


```python
import numpy as np
x=np.array([1,2,3,4,5 ])
y=np.array([2,3,6,8,10])
```

```python
import  matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#  interpolate the data using linear interploation
x_new=np.linspace(1,5,10) # create new x values
y_interp=np.interp(x_new,x,y) # interpolat
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

# cubic interpolation

```python
x=np.array([1,2,3,4,5 ])
y=np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
f=interp1d(x,y, kind='cubic')
```

```python
x_new=np.linspace(1,5,10)
y_interp=f(x_new)
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

poly_interpolation

```python
import numpy as np

# create some sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
```

```python
# interpolate the data using polynomial interpolation
p=np.polyfit(x,y,2)
```

```python
x_new = np.linspace(1, 5, 10) # create new x values
y_interp = np.polyval(p, x_new) # interpolate y values
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

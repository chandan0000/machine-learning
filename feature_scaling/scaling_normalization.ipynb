```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill=list(df['total_bill'])
```

```python
import numpy as np
mean=np.mean(total_bill)
std=np.std(total_bill)
mean, std
```

```python
normalized_data=[]
for i in total_bill:
    z_score=(i-mean)/std
    normalized_data.append(z_score)
normalized_data
```

```python
sns.histplot(normalized_data),
```
